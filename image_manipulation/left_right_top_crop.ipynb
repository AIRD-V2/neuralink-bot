# Define Various Crops 

```python
def crop_lower_half(fnames):
    for i, file in enumerate(fnames):
        cropped_image = crop(open_image(file), 1910, 1.475, 1)
        cropped_image.save("/home/andrew/Github/neuralink-bot/image_manipulation/top_crop/top_crop_%s.jpg" % (str(i)))
```

```python
crop_lower_half(fnames)
```

```python
# Top Crop fnames
top_crops_path = Path("/home/andrew/Github/neuralink-bot/image_manipulation/top_crop")
top_fnames = get_image_files(top_crops_path)
```

```python
def left_crop(fnames):
    for i, file in enumerate(fnames):
        left_image = crop(open_image(file), 955, 1, 0)
        left_image.save("/home/andrew/Github/neuralink-bot/image_manipulation/left_crop/left_crop_%s.jpg" % (str(i)))
```

```python
left_crop(top_fnames)
```

```python
def right_crop(fnames):
    for i, file in enumerate(fnames):
        right_image = crop(open_image(file), 955, 1, 1)
        right_image.save("/home/andrew/Github/neuralink-bot/image_manipulation/right_crop/right_crop_%s.jpg" % (str(i)))
```

```python
right_crop(top_fnames)
```
